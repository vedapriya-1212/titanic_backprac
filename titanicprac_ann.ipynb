```python
# Normalized Inputs
x1 = 0.20   # Pclass
x2 = 0.24   # Age
x3 = 0.80   # Fare

# Target Output
target = 1

# Learning Rate
learning_rate = 0.1
```

```python
# Input Layer -> Hidden Layer Weights

w_x1_h1 = 0.11
w_x2_h1 = 0.14
w_x3_h1 = 0.17

w_x1_h2 = 0.21
w_x2_h2 = 0.24
w_x3_h2 = 0.27

# Hidden Layer Biases
b_h1 = 0.1
b_h2 = 0.1

# Hidden Layer -> Output Layer Weights

w_h1_o1 = 0.31
w_h2_o1 = 0.34

# Output Bias
b_o1 = 0.1
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
# Net input for Hidden Neuron h1
net_h1 = (x1 * w_x1_h1) + (x2 * w_x2_h1) + (x3 * w_x3_h1) + b_h1

# Net input for Hidden Neuron h2
net_h2 = (x1 * w_x1_h2) + (x2 * w_x2_h2) + (x3 * w_x3_h2) + b_h2

print("Net input h1 =", round(net_h1, 4))
print("Net input h2 =", round(net_h2, 4))
```

```python
# Output of hidden neurons

h1 = sigmoid(net_h1)
h2 = sigmoid(net_h2)

print("Output h1 =", round(h1, 4))
print("Output h2 =", round(h2, 4))
```

```python
# Net input at output neuron

net_o1 = (h1 * w_h1_o1) + (h2 * w_h2_o1) + b_o1

# Final predicted output

output = sigmoid(net_o1)

print("Net input output neuron =", round(net_o1, 4))
print("Final predicted output =", round(output, 4))
```

```python
# Mean Squared Error

mse = 0.5 * ((target - output) ** 2)

print("Mean Squared Error =", round(mse, 4))
```

```python
# Output layer gradient

delta_o1 = (target - output) * output * (1 - output)

print("Output layer gradient =", round(delta_o1, 4))
```

```python
# Hidden layer gradients

delta_h1 = h1 * (1 - h1) * (delta_o1 * w_h1_o1)

delta_h2 = h2 * (1 - h2) * (delta_o1 * w_h2_o1)

print("Hidden layer gradient h1 =", round(delta_h1, 4))
print("Hidden layer gradient h2 =", round(delta_h2, 4))
```

```python
# Update weights from hidden layer to output layer

new_w_h1_o1 = w_h1_o1 + learning_rate * delta_o1 * h1

new_w_h2_o1 = w_h2_o1 + learning_rate * delta_o1 * h2

# Update output bias

new_b_o1 = b_o1 + learning_rate * delta_o1

print("Updated w_h1_o1 =", round(new_w_h1_o1, 4))
print("Updated w_h2_o1 =", round(new_w_h2_o1, 4))
print("Updated output bias =", round(new_b_o1, 4))
```

```python
# Update weights connected to h1

new_w_x1_h1 = w_x1_h1 + learning_rate * delta_h1 * x1

new_w_x2_h1 = w_x2_h1 + learning_rate * delta_h1 * x2

new_w_x3_h1 = w_x3_h1 + learning_rate * delta_h1 * x3

# Update bias for h1

new_b_h1 = b_h1 + learning_rate * delta_h1

print("Updated w_x1_h1 =", round(new_w_x1_h1, 4))
print("Updated w_x2_h1 =", round(new_w_x2_h1, 4))
print("Updated w_x3_h1 =", round(new_w_x3_h1, 4))
print("Updated bias h1 =", round(new_b_h1, 4))
```

```python
# Update weights connected to h2

new_w_x1_h2 = w_x1_h2 + learning_rate * delta_h2 * x1

new_w_x2_h2 = w_x2_h2 + learning_rate * delta_h2 * x2

new_w_x3_h2 = w_x3_h2 + learning_rate * delta_h2 * x3

# Update bias for h2

new_b_h2 = b_h2 + learning_rate * delta_h2

print("Updated w_x1_h2 =", round(new_w_x1_h2, 4))
print("Updated w_x2_h2 =", round(new_w_x2_h2, 4))
print("Updated w_x3_h2 =", round(new_w_x3_h2, 4))
print("Updated bias h2 =", round(new_b_h2, 4))
```

```python
print("===== FINAL UPDATED WEIGHTS =====")

print("\nInput -> Hidden Layer")

print("w_x1_h1 =", round(new_w_x1_h1, 4))
print("w_x2_h1 =", round(new_w_x2_h1, 4))
print("w_x3_h1 =", round(new_w_x3_h1, 4))

print("w_x1_h2 =", round(new_w_x1_h2, 4))
print("w_x2_h2 =", round(new_w_x2_h2, 4))
print("w_x3_h2 =", round(new_w_x3_h2, 4))

print("\nHidden -> Output Layer")

print("w_h1_o1 =", round(new_w_h1_o1, 4))
print("w_h2_o1 =", round(new_w_h2_o1, 4))

print("\nUpdated Biases")

print("b_h1 =", round(new_b_h1, 4))
print("b_h2 =", round(new_b_h2, 4))
print("b_o1 =", round(new_b_o1, 4))
```
